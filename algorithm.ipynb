<h6 style="color:gold">Analysis of Sorting Algorithms Used</h6>
In the provided code, three different sorting algorithms are implemented to sort a list of student dictionaries based on different keys: ID, name, and CGPA. The algorithms used are Bubble Sort, Insertion Sort, and Selection Sort. Below is a comparison of their performance, advantages, and disadvantages.
<h6 style="color:gold">1. Bubble Sort</h6> (Used for Sorting by ID)<h6 style="color:gold">Performance:</h6><h6 style="color:gold">Time Complexity:</h6> O(n^2) in the average and worst cases, where n is the number of elements in the list. The best case is O(n) when the list is already sorted.<h6 style="color:gold">Space Complexity: </h6>O(1) since it is an in-place sorting algorithm.
h6 style="color:gold">Advantages:</h6>
Simple to understand and implement.Works well for small datasets or nearly sorted data.<h6 style="color:gold">Disadvantages:</h6>
Inefficient for large datasets due to its quadratic time complexity.
Generally performs worse than other algorithms like Insertion Sort and Selection Sort for most cases.<h6 style="color:gold">2. Insertion Sort </h6>(Used for Sorting by Name)
<h6 style="color:gold">Performance:</h6>
<h6 style="color:gold">Time Complexity:</h6> O(n^2) in the average and worst cases, but O(n) in the best case when the list is already sorted.
Space Complexity: O(1) as it is also an in-place sorting algorithm.
<h6 style="color:gold">Advantages:</h6>
More efficient than Bubble Sort for small datasets or when the list is partially sorted.
Stable sort, meaning that it maintains the relative order of records with equal keys.
<h6 style="color:gold">Disadvantages:</h6>
Still has a quadratic time complexity, making it inefficient for large datasets.
Performance degrades significantly as the size of the dataset increases.
<h6 style="color:gold">3. Selection Sort</h6> (Used for Sorting by CGPA)
<h6 style="color:gold">Performance:</h6>
<h6 style="color:gold">Time Complexity:</h6> O(n^2) for all cases (best, average, and worst).
<h6 style="color:gold">Space Complexity:</h6> O(1) since it is an in-place sorting algorithm.
<h6 style="color:gold">Advantages:</h6>
Simple to implement and understand.
Performs well on small lists and has a predictable performance.
<h6 style="color:gold">Disadvantages:</h6>
Inefficient for large datasets due to its quadratic time complexity.
Not stable, meaning that it may change the relative order of records with equal keys.
<h6 style="color:gold">Summary</h6>
In summary, all three sorting algorithms (Bubble Sort, Insertion Sort, and Selection Sort) have a time complexity of O(n^2) in the average and worst cases, making them inefficient for large datasets. However, they are simple to implement and understand, which can be beneficial for educational purposes or small datasets.
Bubble Sort is the least efficient and is generally not recommended for practical use.
Insertion Sort is more efficient for small or partially sorted datasets and is stable.
Selection Sort is straightforward but not stable and does not offer any performance advantage over the other two for larger datasets.
For practical applications, especially with larger datasets, more efficient algorithms like Merge Sort or Quick Sort should be considered, as they have better average and worst-case time complexities (O(n log n)).
